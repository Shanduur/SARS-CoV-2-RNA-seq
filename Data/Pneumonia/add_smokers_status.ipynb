```python
import pandas as pd

meta = pd.read_csv('GSE164948_covid_control_count_metadata.csv', sep = ';')
meta.head()
```

```python
new_meta = pd.read_csv('smoking_status_elife_97.csv', sep = ';')
new_meta
```

```python
s = new_meta['status'].unique()

new_meta['status_num'] = new_meta['status'].replace(s, [1, 0, -1])
new_meta
```

```python
meta['smoking'] = 0

for id in new_meta['id']:
    status = new_meta[new_meta['id'] == id]['status_num'].tolist()[0]
    meta.loc[meta['id_new'] == id, 'smoking'] = meta.loc[meta['id_new'] == id, 'smoking'].apply(lambda x: status)

meta['smoking'].unique()
```

```python
cols = meta.columns.tolist()
cols[0] = ''
cols
```

```python
import csv


meta.to_csv('GSE164948_covid_control_count_metadata_added.csv', sep=';', 
            index=False, quotechar='"', header=cols, quoting=csv.QUOTE_NONNUMERIC, line_terminator='\r\n')
```
